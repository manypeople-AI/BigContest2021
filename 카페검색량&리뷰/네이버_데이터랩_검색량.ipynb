```python
datalab = datalab.drop([datalab.index[66],datalab.index[65],datalab.index[64],datalab.index[63],datalab.index[62],datalab.index[61],datalab.index[60]])
datalab
```

```python
datalab.rename(columns = {'url':'날짜', 'http://datalab.naver.com/keyword/trendResult.naver?hashKey=N_1d30f943c696fd8949c21e6e01363da2':'군산 카페','Unnamed: 3':'군산 고군산군도(선유도) 카페'}, inplace = True)
datalab
```

```python
datalab.drop(['Unnamed: 2'],axis = 1, inplace = True)
datalab
```

```python
datalab['합'] = 1
datalab
```

```python
datalab.drop(['날짜'],axis = 1, inplace=True)
datalab
```

```python
datalab['군산 카페'] = datalab['군산 카페'].apply(pd.to_numeric)
datalab['군산 고군산군도(선유도) 카페'] = datalab['군산 고군산군도(선유도) 카페'].apply(pd.to_numeric)
```

```python
datalab = datalab.groupby(datalab['합']).sum()
datalab
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
group_names = datalab.columns
group_sizes = datalab.values

wedgeprops={'width': 0.6, 'edgecolor': 'w', 'linewidth': 3}
explode = [0.02, 0.02]



plt.title('수요비중(검색량)',{'fontsize':25})

plt.pie(group_sizes, 
        labels=group_names, 
        colors= [ 'silver','yellowgreen'], 
        autopct='%.1f%%', # second decimal place
        startangle=90,
        textprops={'fontsize': 16},
        explode=explode,
        wedgeprops = wedgeprops) # text font size

plt.axis('equal')
plt.show()
```
